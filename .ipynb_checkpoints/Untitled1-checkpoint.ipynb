```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

#loading images
img1 = cv.imread('image1.jpg',cv.IMREAD_GRAYSCALE)          # queryImage
img2 = cv.imread('image2.jpg',cv.IMREAD_GRAYSCALE) # trainImage
img3 = cv.imread('image3.jpg',cv.IMREAD_GRAYSCALE)

# Initiate SIFT detector
sift = cv.SIFT_create()

# find the keypoints and descriptors with SIFT
kp1, des1 = sift.detectAndCompute(img1,None)
img_1 = np.zeros([(img1.shape[0]+img2.shape[0]),(img1.shape[1]+img2.shape[1])]).astype(np.uint8)
img_1[0:img2.shape[0],img1.shape[1]:(img1.shape[1]+img2.shape[1])] =img2 
kp2, des2 = sift.detectAndCompute(img_1,None)
kp3, des3 = sift.detectAndCompute(img3,None)

FLANN_INDEX_KDTREE = 1
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks = 50)
flann = cv.FlannBasedMatcher(index_params, search_params)
matches1 = flann.knnMatch(des1,des2,k=2)
matches2 = flann.knnMatch(des2,des3,k=2)
# store all the good matches as per Lowe's ratio test.
good = []
for m,n in matches1:
    if m.distance < 0.7*n.distance:
        good.append(m)
plt.imshow(img1,"gray")
plt.show()  
plt.imshow(img_1,"gray")
plt.show()  
plt.imshow(img3,"gray")
plt.show() 
```

```python
MIN_MATCH_COUNT = 10
if len(good)>MIN_MATCH_COUNT:
    src_pts = np.float32([ kp1[m.queryIdx].pt for m in good ]).reshape(-1,1,2)
    dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good ]).reshape(-1,1,2)
    
    M, mask = cv.findHomography(src_pts,dst_pts, cv.RANSAC,5.0)
    print(M)
    
width = img1.shape[1] + img_1.shape[1]
height = img1.shape[0] +img_1.shape[0]
img_2 = np.zeros([img1.shape[0] +img_1.shape[0],img1.shape[1] + img_1.shape[1]]).astype(np.uint8)
img_2[0:img_1.shape[0],0:(img_1.shape[1])] =img_1 
result = cv.warpPerspective(img1,M,(width,height)) 
result = cv.bitwise_or(result,img_2)/2
plt.imshow(result,"gray")
plt.show()
   
```

```python
good1 = []
for m,n in matches2:
    if m.distance < 0.7*n.distance:
        good1.append(m)
        
MIN_MATCH_COUNT = 10
if len(good)>MIN_MATCH_COUNT:
    src_pts = np.float32([ kp2[m.queryIdx].pt for m in good1 ]).reshape(-1,1,2)
    dst_pts = np.float32([ kp3[m.trainIdx].pt for m in good1 ]).reshape(-1,1,2)
    
    M, mask = cv.findHomography(dst_pts,src_pts,cv.RANSAC,5.0)
    print(M)
    
width = img3.shape[1] + result.shape[1]
height = img3.shape[0] + result.shape[0]
img_3 = np.zeros([img3.shape[0] + result.shape[0],img3.shape[1] + result.shape[1]]).astype(np.uint8)
img_3[0:result.shape[0],0:(result.shape[1])] =result 
result2 = cv.warpPerspective(img3,M,(width,height)) 
result2 = cv.bitwise_or(result2,img_3)/4
#result2[0:result.shape[0],0:(result.shape[1])] =result
plt.imshow(result2,"gray")
plt.show()
```
